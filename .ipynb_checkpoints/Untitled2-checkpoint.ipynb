```python
val_input_names = ["/Users/srinathramalingam/Desktop/codebase/TransUnet/dataset/val/" + i for i in os.listdir("/Users/srinathramalingam/Desktop/codebase/TransUnet/dataset/val")]
val_label_names = ["/Users/srinathramalingam/Desktop/codebase/TransUnet/dataset/val_labels/" + i for i in os.listdir("/Users/srinathramalingam/Desktop/codebase/TransUnet/dataset/val")]

import cv2
import numpy as np
import matplotlib.pyplot as plt

def predict(image, model, mask):
    img = cv2.imread(image)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    img = np.array(img)/255
    img = img.reshape((1,256,256,3))
    prediction = model.predict(img)
    prediction = tf.clip_by_value(prediction, 0.0, 1.0)
    pred = prediction * 255
    pred = np.array(pred).reshape((256,256,1))
    img = img.reshape((256,256,3))
    mas = cv2.imread(mask)
    return img, pred, mas
```

```python
for file in range(len(val_input_names)):
    img, pred1, mas = predict(val_input_names[file], network.model, val_label_names[file])
    
    plt.subplot(1, 3, 1)
    plt.imshow(img)

    plt.subplot(1, 3, 2)
    plt.imshow(pred1)
    
    plt.subplot(1, 3, 3)
    plt.imshow(mas)

    plt.show()
```

```python
import os
import argparse
import pandas as pd
import tensorflow as tf
from bp import Environment
import TransUnet.experiments.config as conf
from dTurk.utils.clr_callback import CyclicLR
import TransUnet.models.transunet as transunet
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
from train_helpers import dice_loss, mean_iou, oversampling, create_dataset
from dTurk.models.sm_models.losses import DiceLoss
from focal_loss import BinaryFocalLoss
import os
import argparse
import pandas as pd
import tensorflow as tf
from bp import Environment
from dTurk.utils.clr_callback import CyclicLR
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
from train_helpers import mean_iou, oversampling, create_dataset

from dTurk.models.SM_UNet import SM_UNet_Builder

env = Environment()

config = conf.get_transunet()
config["image_size"] = 256
config["filters"] = 3
config["n_skip"] = 3
config["decoder_channels"] = [128, 64, 32, 16]
config["resnet"]["n_layers"] = (3, 4, 9, 12)
config["dropout"] = 0.1
config["grid"] = (28, 28)
builder = SM_UNet_Builder(
    encoder_name='efficientnetv2-l',
    input_shape=(256, 256, 3),
    num_classes=3,
    activation="softmax",
    train_encoder=False,
    encoder_weights="imagenet",
    decoder_block_type="upsampling",
    head_dropout=0,  # dropout at head
    dropout=0,  # dropout at feature extraction
)
```

```python
network = transunet.TransUnet(config, trainable=False)
model1 = network.model.load_weights("weights/checkpoint/")
```

```python
builder = SM_UNet_Builder(
    encoder_name='efficientnetv2-l',
    input_shape=(256, 256, 3),
    num_classes=3,
    activation="softmax",
    train_encoder=False,
    encoder_weights="imagenet",
    decoder_block_type="upsampling",
    head_dropout=0,  # dropout at head
    dropout=0,  # dropout at feature extraction
)

model = builder.build_model()
```

```python
model2 = model.load_weights("unet/checkpoint/")
```

```python
def predict(image, model1):
    img = cv2.imread(image)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    img = np.array(img)/255
    img = img.reshape((1,256,256,3))
    prediction1 = model1.predict(img)
    prediction1 = tf.clip_by_value(prediction1, 0.0, 1.0)
    prediction2 = model2.predict(img)
    prediction2 = tf.clip_by_value(prediction2, 0.0, 1.0)
    pred1 = prediction1 * 255
    pred1 = np.array(prediction1).reshape((256,256,3))
    pred2 = prediction2 * 255
    pred2 = np.array(prediction2).reshape((256,256,3))
    img = img.reshape((256,256,3))
    return img, pred1, pred2
```

```python
val_input_names = ["/Users/srinathramalingam/Desktop/codebase/Jupyter_predictions_Nearspace/MACH-77-it3/val/" + i for i in os.listdir("/Users/srinathramalingam/Desktop/codebase/Jupyter_predictions_Nearspace/MACH-77-it3/val")]
val_label_names = ["/Users/srinathramalingam/Desktop/codebase/Jupyter_predictions_Nearspace/MACH-77-it3/val_labels/" + i for i in os.listdir("/Users/srinathramalingam/Desktop/codebase/Jupyter_predictions_Nearspace/MACH-77-it3/val")]

import cv2
import numpy as np
import matplotlib.pyplot as plt

def predict(image, model, mask):
    img = cv2.imread(image)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    img = np.array(img)/255
    img = img.reshape((1,256,256,3))
    prediction = model.predict(img)
    prediction = tf.clip_by_value(prediction, 0.0, 1.0)
    pred = np.array(prediction).reshape((256,256,3))
    img = img.reshape((256,256,3))
    mas = cv2.imread(mask)
    mas = cv2.cvtColor(mas, cv2.COLOR_BGR2RGB)
    return img, pred, mas
```

```python
for file in range(len(val_input_names)):
    img, pred1, mas = predict(val_input_names[file], network.model, val_label_names[file])
    
    plt.subplot(1, 3, 1)
    plt.imshow(img)

    plt.subplot(1, 3, 2)
    plt.imshow(pred1)
    
    plt.subplot(1, 3, 3)
    plt.imshow(mas/255)

    plt.show()
```
